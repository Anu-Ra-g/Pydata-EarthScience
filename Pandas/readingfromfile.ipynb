```python
bios_new = bios.copy()
```

```python
bios_new['first_name'] = bios_new['name'].str.split(" ").str[0]
bios_new
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date']) #converting into datetime type
```

```python
bios_new
```

```python
bios_new.info()
```

```python
bios_new['born_year'] = bios_new["born_datetime"].dt.year
bios_new.head()
```

```python
bios['height_category'] = bios["height_cm"].apply(lambda x: 'Short' if x < 165 else 'Average' if x < 185 else 'Tall')
bios
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'
    
bios['Category'] = bios.apply(categorize_athlete, axis=1)
bios
# In pandas, when you use the apply() function with axis=0, it applies the function along the columns of the DataFrame. This means the function (categorize_athlete in your case) would receive each column as a Series, rather than each row as a Series
```

## Advanced Functionalities

```python
coffee['pct change'] = coffee['revenue'] /coffee['yesterday_revenue'] * 100
```

```python
coffee
```

```python
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
```

```python
bios.sort_values(['height_rank'], ascending=False)
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum() # sums 3 values in the "Units Sold" column
```
